```python
import numpy as np

import pandas as pd

# Cargar los datos desde archivos CSV
players = pd.read_csv('players.csv')

def formacion_mas_cara_actual(n_defender, n_midfield, n_attack):
  return  pd.concat([
      players[players['position'] == 'Attack'].nlargest(n_attack, 'market_value_in_eur'),
      players[players['position'] == 'Midfield'].nlargest(n_midfield, 'market_value_in_eur'),
      players[players['position'] == 'Defender'].nlargest(n_defender, 'market_value_in_eur'),
      players[players['position'] == 'Goalkeeper'].nlargest(1, 'market_value_in_eur')
      ])
```

```python
print(formacion_mas_cara_actual(4, 3, 3))
```

```python
def select_top_players(position, top_n):
    return players[players['position'] == position].nlargest(top_n, 'market_value_in_eur')

def select_historic_top_players(position, top_n):
    return players[players['position'] == position].nlargest(top_n, 'highest_market_value_in_eur')

# Seleccionar los 3 delanteros más caros
top_forwards = select_top_players('Attack', 3)

# Seleccionar los 3 mediocampistas más caros
top_midfielders = select_top_players('Midfield', 3)

# Seleccionar los 4 defensores más caros
top_defenders = select_top_players('Defender', 4)

top_goalkeeper = select_top_players('Goalkeeper', 1)


# Seleccionar los 3 delanteros más caros
top_historic_forwards = select_historic_top_players('Attack', 3)

# Seleccionar los 3 mediocampistas más caros
top_historic_midfielders = select_historic_top_players('Midfield', 3)

# Seleccionar los 4 defensores más caros
top_historic_defenders = select_historic_top_players('Defender', 4)

top_historic_goalkeeper = select_historic_top_players('Goalkeeper', 1)
# Concatenar todos los jugadores seleccionados para formar la plantilla
starting_eleven = pd.concat([top_forwards, top_midfielders, top_defenders, top_goalkeeper])
starting_historic_eleven = pd.concat([top_historic_forwards, top_historic_midfielders, top_historic_defenders, top_historic_goalkeeper])


print("Plantilla 4-3-3 con los jugadores más caros en cada posición:")
print(starting_eleven.to_string(index=False))
print("Historicos:")
print(starting_historic_eleven.to_string(index=False))
```
